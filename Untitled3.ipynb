```python
import pandas as pd

##читаем таблицу
df = pd.read_csv('sample_data/california_housing_train.csv') ##указываем в качестве доп.аргумента sep= и разделитель, которым разделяются данные в файле
```

```python
##Изобразите отношение households к population с помощью точечного графика
import seaborn as sns
sns.scatterplot(data=df, x="households", y="population")
```

```python
##Визуализировать longitude по отношению к median_house_value, используя линейный график
sns.relplot(x="longitude", y="median_house_value", kind="line", data=df)
```

```python
##Представить гистограмму по housing_median_age
sns.histplot(data=df, x="housing_median_age")
```

```python
##Изобразить гистограмму по median_house_value с оттенком housing_median_age
df.loc[df['housing_median_age'] < 15, 'housing_age'] = 'new'
df.loc[(df['housing_median_age'] >= 15) & (df['housing_median_age'] <= 30), 'housing_age'] = 'medium'
df.loc[df['housing_median_age'] > 30, 'housing_age'] = 'old'

sns.displot(df, x="median_house_value", hue="housing_age")

```
